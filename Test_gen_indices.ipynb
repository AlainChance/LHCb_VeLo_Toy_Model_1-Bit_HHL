# 1-Bit HHL track simulation toy model - Function gen_indices()
## Define function gen_indices()
Input parameters:
- `layers`: number of layers, default: 3
- `npart` : number of particles, default: 2

Returns a list of segment indices ordered as shown in the left panel of Figure 2. of the following document:
[ALGO-3] [Xenofon Chiotopoulos, TrackHHL: A Quantum Computing Algorithm for Track Reconstruction at the LHCb](https://cds.cern.ch/record/2950969/files/document.pdf)

The function `setup_Hamiltonian()` calls `gen_indices()`:
```
npart = sum(param["n_particles"])
segment_indices = self.gen_indices(npart=npart, layers=3)
```
For a detector with $l$ layers and $m$ reconstructible particle tracks, the system size scales as $N = m^2(l-1)$ [ALGO-2](https://doi.org/10.48550/arXiv.2601.07766).

Author: Alain Chancé

````python
def gen_indices(layers=3, npart=2):
    indices = []
    for j in range(layers-1):
        for i in range(npart):
            indices.append(i * (npart+1) + j*npart**2)
    return indices
````

## Generate list of indices for 2 to 8 particles and 3 layers

````python
print(f"Number of particles   Active segment indices")
for npart in range(2,9):
    segment_indices = gen_indices(layers=3, npart=npart, )
    print(f"{npart}                     {segment_indices}")
````
